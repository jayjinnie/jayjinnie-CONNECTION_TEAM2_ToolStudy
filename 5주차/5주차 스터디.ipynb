# 기초통계 실습

```python
# 한국 전체 고등학생 전체의 시험 성적 평균은 18이고, 표준편차는 3입니다.
# 연구팀이 개발한 신약이 고등학생의 시험 성적에 영향을 주는지 알아봅시다.
# 신약을 복용한 30명의 시험 성적 평균은 19.3이었을 때, 주어진 정보를 가지고 가설 검정을 진행해봅시다.


# 모집단: population
# 표본집단: sample
pop_mean = 18
pop_sd = 3
sample_mean = 19.3
sample_n = 30

# 귀무가설
H0 = "신약을 복용한 사람의 성적 평균은 18이다. (신약은 시험성적에 영향을 주지 않는다)"
# 대립가설
H1 = "신약을 복용한 사람의 성적 평균은 18이 아니다. (신약은 시험성적에 영향을 준다)"

print("귀무가설은", H0)
print("대립가설은", H1)
```

```python
import numpy as np

# 유의수준 0.5
# 검정통계량 계산
# 검정통계량 Z = (표본평균-모평균) / (모표준편차/(루트n))

sample_error = sample_mean - pop_mean
SE = pop_sd / np.sqrt(sample_n)

Z = sample_error / SE
print("검정통계량 Z는", round(Z,1))
```

```python
from scipy import stats
import matplotlib.pyplot as plt

# 평균 0, 표준편차 1
z_dist = stats.norm(loc = 0, scale = 1)
x = np.linspace(-10, 10, 100)
y = z_dist.pdf(x)

plt.plot(x,y)
plt.title("PDF of z-distribution")
```

```python
z_dist = stats.norm(loc = 0, scale = 1)
# 간격 설정
x = np.linspace(-10, 10, 100)
y = z_dist.pdf(x)

plt.plot(x,y)
plt.title("PDF of z-distribution")

# 오른쪽
right = np.linspace(abs(Z), x[-1], 50)
plt.fill_between(right, z_dist.pdf(right), color = "red")

# 왼쪽
left = np.linspace(-x[-1], -abs(Z), 50)
plt.fill_between(left, z_dist.pdf(left), color = "red")
```

```python
# p value와 유의수준을 비교하여 가설 기각 여부 판단
# sf는 괄호 안 값보다 큰 영역(오른쪽)의 값

# 임계치와 검정 통계량 Z값 비교하여 가설 기각 여부 판단
if abs(Z) <= 1.96:
    print("귀무가설", H0, "기각 실패")
    print("대립가설", H1, "채택 X")
else:
    print("귀무가설", H0, "기각 성공")
    print("대립가설", H1, "채택")
```
